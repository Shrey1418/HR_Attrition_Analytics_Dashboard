```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Excel Database.csv')
```

```python
df.shape
```

```python
df.info()
```

```python
df.head(10)
```

```python
df.isnull().sum()
```

```python
df.drop('EmployeeCount', axis = 1, inplace = True)
```

```python
df.drop('Over18', axis = 1, inplace = True)
```

```python
df.drop('StandardHours', axis = 1, inplace = True)
```

```python
df.drop('EmployeeNumber', axis = 1, inplace = True)
```

```python
df.duplicated().sum()
```

```python
df['Attrition'] = df['Attrition'].map({
    'Yes' : 1,
     'No' : 0
})
```

```python
df.head(5)
```

```python
df['AgeGroup'] = pd.cut(
    df['Age'],
    bins = [18,30,40,50,60],
    labels = ['18-30','31-40','41-50','51-60']
)
```

```python
df['ExperienceLevel'] = pd.cut(
    df['TotalWorkingYears'],
    bins = [0,3,7,15,40],
    labels = ['Junior','MidLevel','Senior','Expert']
)
```

```python
df.head(5)
```

```python
df['TenureGroup'] = pd.cut(
    df['YearsAtCompany'],
    bins = [-1,2,5,10,40],
    labels = [''
    '0-2 Years',
    '3-5 Years',
    '6-10 Years',
    '10+ Years']
)
```

```python
df[['YearsAtCompany','TenureGroup']].head(10)
```

```python
df['SalarySlab'] = pd.cut(
    df['MonthlyIncome'],
    bins = [0,5000,10000,20000],
    labels = [''
    'Low Salary',
    'Medium Salary',
    'High Salary',
    ]
)
```

```python
df[['SalarySlab','MonthlyIncome']].head(10)
```

```python
attrition_rate = df['Attrition'].mean() * 100 
print(f"Overall Attrition Rate : {attrition_rate:.2f}%")
```

```python
df['Attrition'].value_counts()
```

```python
dept_attrition = (
    df.groupby('Department')['Attrition'].mean() * 100
)
print(dept_attrition.sort_values(ascending = False))
```

```python
dept_attrition.sort_values().plot(
    kind = 'barh'
)
plt.title('Attrition by Department')
plt.xlabel('Attrition Perentage')
plt.show()
```

### Observation : Sales Department has the highest Attrition
### Business Meaning : Sales employees may be experiencing higher workload pressure, target stress, or lower job satisfaction
### Recommendation : Investigate workload, incentives, and manager quality in Sales

```python
role_attrition = (
    df.groupby('JobRole')['Attrition'].mean() * 100
)
print(role_attrition.sort_values(ascending = False))
```

```python
role_attrition.sort_values().plot(
    kind = 'barh'
)
plt.title('Attrition by Job Role')
plt.xlabel('Attrition Perentage')
plt.show()
```

### Observation : Sales Representatives show the highest attrition rate among all job roles, while managerial and director-level roles have the lowest attrition.

### Business Meaning : Frontline sales employees may be facing higher work pressure, strict targets, frequent client interactions, or lower job satisfaction, leading to increased employee turnover. Higher-level roles may experience lower attrition due to better compensation, stability, and career growth opportunities.

### Recommendation : The company should improve employee engagement, incentives, workload management, and career growth opportunities for high-risk job roles, especially Sales Representatives and Laboratory Technicians.

```python
overtime_attrition = (
    df.groupby('OverTime')['Attrition'].mean() * 100
)
print(overtime_attrition)
```

```python
sns.barplot(
    x = 'OverTime',
    y = 'Attrition',
    data = df
)
plt.title('Overtime vs Attrition')
plt.ylabel('Attrition Rate')
plt.show()
```

### Observation : Employees working overtime have significantly higher attrition rates compared to employees who do not work overtime.

### Business Meaning : Excessive workload and longer working hours may be causing stress, burnout, and poor work-life balance, leading employees to leave the company.

### Recommendation : The company should monitor overtime frequency, improve workload distribution, and introduce work-life balance initiatives to reduce employee burnout and attrition.

```python
sns.barplot(
    x = 'Attrition',
    y = 'MonthlyIncome',
    data = df
)
plt.title('MonthlyIncome vs Attrition')
plt.show()
```

### Observation : Employees who left the company had lower average monthly income compared to employees who stayed.

### Business Meaning : Lower compensation may be contributing to employee dissatisfaction and increased attrition, especially among lower-income employees.

### Recommendation : The company should review compensation structures, provide competitive salary packages, and introduce retention incentives for lower-paid employees.

```python
sns.barplot(
    x = 'JobSatisfaction',
    y = 'Attrition',
    data = df
)
plt.title('MonthlyIncome vs Attrition')
plt.show()
```

### Observation : Employees with lower job satisfaction levels show higher attrition rates, while employees with higher job satisfaction tend to stay longer in the company.

### Business Meaning : Low job satisfaction may indicate issues related to work environment, management, workload, career growth, or employee engagement, which can increase employee turnover.

### Recommendation : The company should improve employee engagement programs, provide better career development opportunities, strengthen manager support, and regularly monitor employee satisfaction levels.

```python
sns.barplot(
    x = 'WorkLifeBalance',
    y = 'Attrition',
    data = df
)
plt.title('MonthlyIncome vs Attrition')
plt.ylabel('Attrintion Rate')
plt.show()
```

### Observation : Employees with poor work-life balance (level 1) show the highest attrition rate, while employees with better work-life balance generally show lower attrition.

### Business Meaning : Poor work-life balance may be increasing employee stress, burnout, and dissatisfaction, leading to higher employee turnover.

### Recommendation : The company should improve work-life balance policies by reducing excessive workload, managing overtime effectively, and promoting flexible and employee-friendly work practices.

```python
age_attrition = (
    df.groupby('AgeGroup')['Attrition'].mean() * 100
)
print(age_attrition)
```

```python
age_attrition.plot(
    kind = 'bar'
)
plt.title('Attrintion by Age Group')
plt.ylabel('Attrintion Percentage')
plt.show()
```

### Observation : Employees in the 18–30 age group show the highest attrition rate compared to other age groups.

### Business Meaning : Younger employees may be more likely to switch jobs for better salary, career growth, learning opportunities, or work environment. Early-career employees generally have lower organizational attachment and higher job mobility.

### Recommendation : The company should focus on improving career development programs, mentorship opportunities, learning initiatives, and employee engagement strategies for younger employees to improve retention.

```python
exp_attrition = (
    df.groupby('ExperienceLevel')['Attrition'].mean() * 100
)
print(exp_attrition)
```

```python
exp_attrition.plot(
    kind = 'bar'
)
plt.title('Attrintion by Experience Level')
plt.ylabel('Attrintion Percentage')
plt.show()
```

### Observation : Junior employees show the highest attrition rate, while experienced and expert-level employees show significantly lower attrition.

### Business Meaning : Less experienced employees may be leaving due to limited career growth, lower compensation, lack of job stability, or better external opportunities. Experienced employees are generally more stable because of higher salaries, stronger organizational attachment, and career progression.

### Recommendation : The company should strengthen onboarding, mentorship, career development, and retention programs for junior employees to improve long-term employee retention.

```python
plt.figure(figsize=(14,10))

sns.heatmap(
    df.corr(numeric_only = True),
    annot = True,
    cmap = 'coolwarm'
)
plt.title('Correlation Heatmap')
plt.show()
```

### Observation : Attrition shows negative correlation with factors such as Age, MonthlyIncome, TotalWorkingYears, JobLevel, and YearsAtCompany. This indicates that employees who are older, more experienced, and higher paid are less likely to leave the company.

### Business Meaning : Employees with lower experience, lower salary, and shorter tenure are more likely to leave the organization. Higher compensation, career stability, and long-term organizational attachment appear to improve employee retention.

### Recommendation : The company should focus on retaining early-career and lower-income employees through better compensation, career growth opportunities, mentorship programs, and employee engagement initiatives.
